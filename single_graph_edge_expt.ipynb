A new graph is generated for each edge flip and the spectral distance from the original is computed based on a chosen graph matrix.
```python
import numpy as np
from scipy.linalg import expm, solve_continuous_lyapunov
from scipy.linalg import null_space
import networkx as nx

n = 5
G = nx.erdos_renyi_graph(n, 0.3)

neg_L = -nx.laplacian_matrix(G)
neg_L = neg_L.todense()

E = np.linalg.eigvalsh(neg_L)
print(f"Eigenvalues: {[float(f'{val:.3f}') for val in E]}", end="\n\n")

# method 1
t = 1e6
J1 = expm(neg_L*t)
# print(f"J: {J1}", end="\n\n")

# method 2
V0 = null_space(neg_L)
J2 = V0 @ np.linalg.pinv(V0.T @ V0) @ V0.T
# print(f"J2: {J2}", end="\n\n")

# comparison of J1 and J3
print(f"norm(J1): {np.linalg.norm(J1):.3g}")
print(f"norm(J1 - J2): {np.linalg.norm(J1 - J2):.3g}", end="\n\n")

B = np.eye(n)

Qc = (np.eye(n) - J1) @ B @ B.T @ (np.eye(n) - J1).T
P = solve_continuous_lyapunov(neg_L, Qc)
print(f"P: {P}", end="\n\n")
print(f"J*P*J^T: {J1 @ P @ J1.T}", end="\n\n")
```

```python
import cvxpy as cp

tol = 1e-6
A = neg_L
print(f"eigvals(A) = {np.linalg.eigvals(A)}")
J = J1
n = A.shape[0]
P = cp.Variable((n, n), symmetric=True)

I = np.eye(n)
# Qc = (I-J) @ B @ B.T @ (I-J).T

constraints = [
    A @ P + P @ A.T + Qc == 0,   # Lyapunov equality
    J @ P @ J.T == 0,            # nullspace constraint
    P >> tol                       # optional: P is PSD
]

prob = cp.Problem(cp.Minimize(0), constraints)
# prob.solve(solver=cp.GUROBI)   # or solver=cp.CVXOPT etc.
prob.solve(solver=cp.SCS)   # or solver=cp.CVXOPT etc.
P_opt = P.value

from modules.control import pseudo_gramian_for_semistable_A_inf_horizon

P_opt = pseudo_gramian_for_semistable_A_inf_horizon(A, B, tol)

print(f"P_opt = {P_opt}", end="\n\n")
print(f"norm(J @ P_opt @ J.T) = {np.linalg.norm(J @ P_opt @ J.T):.3g}")
print(f"norm(A @ P_opt + P_opt @ A.T + Qc) = {np.linalg.norm(A @ P_opt + P_opt @ A.T + Qc):.3g}")
print(f"P_opt >> 0: {np.all(np.linalg.eigvals(P_opt) > -tol)}")

if np.linalg.norm(J @ P_opt @ J.T) > tol or np.linalg.norm(A @ P_opt + P_opt @ A.T + Qc) > tol or not np.all(np.linalg.eigvals(P_opt) > -tol):
    raise ValueError("Problem with P_opt")
```
